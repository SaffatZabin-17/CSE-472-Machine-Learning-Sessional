```python
def find_columns_with_question_mark(dataframe):
    columns_with_question_mark = []

    for column in dataframe.columns:
        if dataframe[column].astype(str).str.strip().eq('?').any():
            columns_with_question_mark.append(column)

    return columns_with_question_mark

columns_with_question_mark = find_columns_with_question_mark(dataframe_train)
print(columns_with_question_mark)
```

```python
dataframe_train.replace(to_replace=r'^\s*\?\s*$', value=np.nan, regex=True, inplace=True)

columns_with_question_mark = find_columns_with_question_mark(dataframe_train)

print(columns_with_question_mark)
```

```python
def replace_nan(dataframe):
    for column in dataframe.columns:
        if dataframe[column].isnull().any():
            if pd.api.types.is_numeric_dtype(dataframe[column]):
                mean_value = dataframe[column].mean()
                dataframe[column].fillna(mean_value, inplace=True)
            else: 
                mode_value = dataframe[column].mode()[0]
                dataframe[column].fillna(mode_value, inplace=True)

    return dataframe

dataframe_train = replace_nan(dataframe_train)

dataframe_train.head(100)
```

```python
columns_with_question_mark = find_columns_with_question_mark(dataframe_test)

dataframe_test.replace(to_replace=r'^\s*\?\s*$', value=np.nan, regex=True, inplace=True)

columns_with_question_mark = find_columns_with_question_mark(dataframe_test)

print(columns_with_question_mark)
```

```python
label_encoding_columns_features_train = []
one_hot_encoding_columns_features_train = []
binary_columns_features_train = []

for column in features_train:
    if features_train[column].dtype == object:
        print(column)
        unique_values = features_train[column].unique()
        unique_count = features_train[column].nunique()
        if unique_count == 2:
            if features_train[column].dropna().isin([0, 1]).all():
                binary_columns_features_train.append(column)
            else:
                label_encoding_columns_features_train.append(column)
        else :
            one_hot_encoding_columns_features_train.append(column)
        print(unique_values)
        print(unique_count)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for column in label_encoding_columns_features_train:
    features_train[column] = encoder.fit_transform(features_train[column])

features_train.head(100)
```

```python
features_train_encoded = pd.get_dummies(features_train, columns=one_hot_encoding_columns_features_train).astype(int)
```

```python
one_hot_encoded_columns_features_train = []

for column in features_train_encoded.columns:
    if features_train_encoded[column].nunique() == 2 and column not in (label_encoding_columns_features_train + binary_columns_features_train):
        if features_train_encoded[column].between(0, 1).all():
            one_hot_encoded_columns_features_train.append(column)

one_hot_encoded_columns_features_train
```

```python
excluded_columns = one_hot_encoded_columns_features_train + label_encoding_columns_features_train + binary_columns_features_train

columns_to_scale = [col for col in features_train_encoded.columns if col not in excluded_columns]

columns_to_scale
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def scaling(df, scaling_type):
    if scaling_type == 0:
        scaler = StandardScaler()
    else:
        scaler = MinMaxScaler()

    features_scale = scaler.fit_transform(df[columns_to_scale])

    return features_scale

features_scale = scaling(features_train_encoded, 1)

features_train_encoded[columns_to_scale] = features_scale

features_train_encoded
```

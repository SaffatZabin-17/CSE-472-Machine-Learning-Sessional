# Task 3: Creation of input and output features

```python
features = dataframe.drop('Attrition', axis=1)
target = dataframe['Attrition']
```

# Task 4: Conversion of features into numeric values

```python
for column in features:
    if features[column].dtype == object:
        print(column)
        unique_values = features[column].unique()
        unique_count = features[column].nunique()
        print(unique_values)
        print(unique_count)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
target = encoder.fit_transform(target)
target
```

```python
columns = ['Gender', 'Over18', 'OverTime']

for column in columns:
    features[column] = encoder.fit_transform(features[column])

features
```

```python
#One-hot encodding

columns_one_hot = features.select_dtypes(include=['object']).columns

features_encoded = pd.get_dummies(features).astype(int)

features_encoded
```

```python
columns
```

# Task 5: Scaling of the features

```python
#one_hot_encoded_columns = [col for col in features.columns if features[col].nunique() == 2]

one_hot_encoded_columns = []

for column in features_encoded.columns:
    if features_encoded[column].nunique() == 2 and column not in columns:
        if features_encoded[column].between(0, 1).all():
            one_hot_encoded_columns.append(column)

one_hot_encoded_columns
```

```python
excluded_columns = columns + one_hot_encoded_columns

columns_to_scale = [col for col in features_encoded.columns if col not in excluded_columns]

columns_to_scale
```

```python
#Scaling the data
from sklearn.preprocessing import StandardScaler, MinMaxScaler

def scaling(df, scaling_type):
    if scaling_type == 0:
        scaler = StandardScaler()
    else:
        scaler = MinMaxScaler()

    features_scale = scaler.fit_transform(df[columns_to_scale])

    return features_scale

features_scale = scaling(features_encoded, 1)

features_encoded[columns_to_scale] = features_scale

features_encoded
```

```python
#Correlation analysis

features = pd.DataFrame(features_encoded, columns=features_encoded.columns)
target = pd.DataFrame(target, columns=['Attrition'])
```

```python
features
```

```python
target
```

# Task 6: Correlation analysis

```python
target_series = target['Attrition']

correlations = features.corrwith(target_series)

correlations
```

```python
correlation_matrix = features.corr()

correlation_matrix
```

```python
sorted_correlations = correlations.abs().sort_values(ascending=False)

top_correlations = sorted_correlations.head(20)

top_correlations
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_correlation_matrix = features[top_correlations.index].corr()

plt.figure(figsize=(10,8))
sns.heatmap(top_correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('correlation Matrix')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

class_0 = features.loc[target["Attrition"] == 0]
class_1 = features.loc[target["Attrition"] == 1]

for feature in top_correlations.index:
    plt.plot(class_0[feature], np.zeros_like(class_0[feature]), 'o', label='Class 0')
    plt.plot(class_1[feature], np.zeros_like(class_1[feature]), 'o', label='Class 1')
    #plt.plot(class_2["sepal_width"], np.zeros_like(class_2['sepal_width']), 'o', label='Class 2')

    plt.legend()
    plt.xlabel(feature)
    plt.title('1D Scatter Plot of OverTime by Numeric Classes')
    plt.show()
```

# Bonus Task: Validating the pipeline
